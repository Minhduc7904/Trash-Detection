```python
import os
from PIL import Image
```

```python
def check_images(image_folder):
    """
    Kiểm tra tất cả hình ảnh trong thư mục
    - Kiểm tra file có mở được không
    - Kiểm tra kích thước
    - Loại bỏ file lỗi
    """
    corrupted = []
    for img_file in os.listdir(image_folder):
        if img_file.endswith(('.jpg', 'jpeg', 'png')):
            img_path = os.path.join(image_folder, img_file)
            try:
                img = Image.open(img_path)
                img.verify()
                print(f"✓ {img_file}: {img.size}")
            except Exception as e:
                print(f"✗ {img_file}: Lỗi - {e}")
                corrupted.append(img_file)

    return corrupted
```
